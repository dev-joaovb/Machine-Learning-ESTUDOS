```python
#Divisao dos dados
from sklearn.model_selection import train_test_split

X = train.drop(['Transported'], axis = "columns")
y = train.Transported

column_names_train = train.drop(['Transported'], axis = "columns").columns
column_name_target = train[['Transported']].columns

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 13)
```

```python
train.info()
```

```python
non_numeric_columns = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
train[non_numeric_columns].agg(['unique'])
```

```python
#Usando o One Hot Encoder das colunas string
encoder_transformer = make_column_transformer(
    (OneHotEncoder(), ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP'])
)

encoder_transformer.fit(X_train)

encoder_transformer.transform(X_train)
```

```python
#encoder de todas as variaveis não numericas
from sklearn.compose import make_column_selector

encoder_transformer = make_column_transformer(
    (OneHotEncoder(), make_column_selector(dtype_exclude = np.number)),
    remainder = 'passthrough'
)

encoder_transformer.fit(X_train)

X_train_encoded = encoder_transformer.transform(X_train)
X_train_encoded
```

```python
#utilização de média incondicional para preencher os valores ausentes.
from sklearn.compose import make_column_selector

encoder_imputer_transformer = make_column_transformer(
    (OneHotEncoder(handle_unknown = 'ignore'), make_column_selector(dtype_exclude = np.number)),
    (SimpleImputer(missing_values=np.nan, strategy = 'mean'), make_column_selector(dtype_include = np.number)),
    remainder = 'passthrough'
)

encoder_imputer_transformer.fit(X_train)

X_train_encoded_imputed = encoder_imputer_transformer.fit_transform(X_train)
X_train_encoded_imputed
```

```python
#pre_processamento de test
X_test_encoded_imputed = encoder_imputer_transformer.transform(X_test)
X_test_encoded_imputed
```

```python
from sklearn.neighbors import KNeighborsClassifier

def train_knn_classifier(X_train, y_train, k):
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(X_train, y_train)
  return knn

trained_knn = train_knn_classifier(X_train_encoded_imputed, y_train, 3)
  
y_pred = trained_knn.predict(X_test_encoded_imputed)
y_pred
```

```python
#primeira tentativa de testar o desempenho
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_true = y_test, 
                          y_pred = y_pred)

accuracy
```

```python
#Segunda tentativa
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

encoder_imputer_scaled_transformer = make_column_transformer(
    (make_pipeline(
        (SimpleImputer(strategy = 'mean')),
        (StandardScaler())
    ) , make_column_selector(dtype_include = np.number)),
    (make_pipeline(
        SimpleImputer(strategy = 'most_frequent'),
        OneHotEncoder(handle_unknown = 'ignore'),
    ), make_column_selector(dtype_exclude = np.number)),
    remainder = 'passthrough'
)


encoder_imputer_scaled_transformer.fit(X_train)

X_train_encoded_imputed_scaled = encoder_imputer_scaled_transformer.transform(X_train)
X_test_encoded_imputed_scaled = encoder_imputer_scaled_transformer.transform(X_test)
```

```python
trained_knn2 = train_knn_classifier(X_train_encoded_imputed_scaled, y_train, 3)

y_pred2 = trained_knn2.predict(X_test_encoded_imputed_scaled)

accuracy = accuracy_score(y_true = y_test, 
                          y_pred = y_pred2)

accuracy
```
